```python
from scipy.spatial import distance 
import math
#above we did import some packages in order to find the euclidian the distance.
# here we will find the euclidian distance distance between the class attribute-usd_pledged_real and usd_goal_real.
point1 = np.array((kickstart.usd_pledged_real))
point2 = np.array((kickstart.usd_goal_real)) 
dist = np.linalg.norm(point1 - point2)

print("the distance between usd_pledged_real and usd_goal_real is: ", dist)
```

```python
# here we will find the euclidian distance distance between the class attribute-usd_pledged_real and backers.
point1 = np.array((kickstart.usd_pledged_real))
point2 = np.array((kickstart.backers)) 
dist = np.linalg.norm(point1 - point2)

print("the distance between usd_pledged_real and bakers is: ", dist)
```
