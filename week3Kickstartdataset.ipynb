#As per the first question we are here doing the data transformation for 3 variables.

```python
# First Variable: state 
# First, we will back the dataframe up.
k1=kickstart

# Now we will build the column then copy the original values into it
k1['state_num'] = k1.state

# Third, we  will convert the  state_num into categorical
k1['state_num']= k1['state_num'].astype('category')

# Then, we  will replace the strings with the category indices
k1['state_num']= k1['state_num'].cat.codes

# Now, we will check the data types
k1.dtypes[['state','state_num']]
#k1.state_num.iloc[20:25]
```

```python
# Second Variable: category
# First, we  will back the dataframe up.
k1=kickstart

# Now we can build the column and copy the original values into it.
k1['category_num'] = k1.state

# Third, we  will now convert category_num into categorical
k1['category_num']= k1['category_num'].astype('category')

# Then, we  can replace the strings with the category indices
k1['category_num']= k1['category_num'].cat.codes

# Now, we  will check the data types
k1.dtypes[['category','category_num']]

#k1[['category', 'category_num']].iloc[20:25]
#k1.category_num.iloc[20:25]
```

```python
# Third Variable: pledged
# First, we back the dataframe up.
k1=kickstart

# Now we build the column and copy the original values into it
k1['pledged_int'] = k1.pledged

# Third, we convert pledged_num into int32
k1['pledged_int']= k1['pledged_int'].astype('int32')

# Then, we replace the strings with the category indices
#k1['category_num']= k1['category_num'].cat.codes

# Now, we  will check the data types
k1.dtypes[['pledged','pledged_int']]
```

```python
bins = np.linspace(kickstart.backers.min(),kickstart.backers.max(),7)
bins
```

```python
labels = ['very_low', 'low', 'medium','Moderate_high' ,'high','very_high']
```

```python
# Variable one: The backers attribute can be  simplified by using Binning methods.
k3 = kickstart
k3['bins'] = pd.cut(k3['backers'], bins=bins, labels=labels, include_lowest=True)
k3[k3.backers>= 40000]
```

```python
k3.groupby(['bins']).count()
```

```python
kickstart.usd_goal_real.describe()
```

```python
# second Variable : usd_goal_real is simplified by using the Bucketing method.
# Setting up our 5 buckets and their lables too.
buckets = [ 0, 30279173.98, 60558347.96, 90837521.94, 121116695.9 ,151395869.9]
bucketlabels = ['very_low', 'low', 'middle', 'high', 'super-high']

# we will use the cut() to separate data into the buckets that we have built.
k3['buckets'] = pd.cut(k3['usd_goal_real'] , bins=buckets, labels=bucketlabels, include_lowest=True)

# lets check  if we have new attributes 'buckets' at the end of the dataset here.
k3[k3.usd_goal_real>= 70000000]
```

```python
k3.groupby(['buckets']).count()
```
